```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.images
```

```python
(digits.data.shape, digits.target.shape, digits.images.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:4], 
                                           digits.target[0:4])):
    plt.subplot(1, 4, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.Dark2)
    plt.title('Training: %i\n' % label, fontsize = 20);
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, 
                                                    digits.target,
                                                   test_size=0.25,
                                                   random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(fit_intercept=True,
                        multi_class='auto',
                        penalty='l2', #ridge regression
                        solver='saga',
                        max_iter=10000,
                        C=50)
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.predict(X_test[0:1]))
print(y_test[0:1])
```

```python
plt.imshow(X_test[0:1].reshape(8,8))
```

```python
clf.predict([X_test[1]])
```
